```python
# categorical data
categorical_df = df.select_dtypes(include=['object', 'category'])
categorical_df["gender"] = df["gender"]
categorical_df["hypertension"] = df["hypertension"]
categorical_df["heart_disease"] = df["heart_disease"]
categorical_df["ever_married"] = df["ever_married"]
categorical_df["work_type"] = df["work_type"]
categorical_df["Residence_type"] = df["Residence_type"]
categorical_df["smoking_status"] = df["smoking_status"]
categorical_df = categorical_df[["gender","hypertension","heart_disease","ever_married","work_type","Residence_type","smoking_status"]]
categorical_df
```

```python
for column in categorical_df.columns:
    print(f"Value counts for {column}:")
    print(categorical_df[column].value_counts())
    print('---' * 10)
```

```python
for i in categorical_df.columns:
  counts = categorical_df[i].value_counts()

  # Plotting
  fig, axes = plt.subplots(1, 2, figsize=(12, 6))

  # Pie chart
  axes[0].pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=140)
  axes[0].set_title(i)

  # Bar chart
  counts.plot(kind='bar', ax=axes[1])
  axes[1].set_title(i)
  axes[1].set_ylabel('Count')
  axes[1].set_xlabel(i)

  plt.tight_layout()
  plt.show()
```

```python
# Create a crosstab to get the counts of each combination
import plotly.express as px
ct = pd.crosstab(df['gender'], df['stroke'])

# Reset the index to make 'gender' a regular column
ct.reset_index(inplace=True)

# Melt the DataFrame to make it suitable for plotting
melted_ct = pd.melt(ct, id_vars='gender', value_vars=[0, 1],
                    var_name='stroke', value_name='Count')

# Create the interactive bar plot
fig = px.bar(melted_ct, x='gender', y='Count',
             color='stroke', title='stroke by gender',
             labels={'gender': 'gender', 'Count': 'Number of People suffering from a stroke'},
             text='Count')

# Show the plot
fig.show()
fig.write_html("stroke_by_gender.html")
```

```python
# The number of women who suffer a stroke is greater than men
# but in similar proportions where approximately 4.71% of females and 5.11% of males in the dataset are suffering from stroke
```
